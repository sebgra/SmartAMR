# Virus selection

```python
import pandas as pd
import numpy as np
import random
```

```python
random.seed(42)
```

```python
N_PHAGES = 100
```

```python
metadata = pd.read_csv("/mnt/data/data/inphared/14Apr2025_data_excluding_refseq.tsv", sep="\t")
```

```python
metadata.head()
```

```python
bacterias = pd.read_csv("../config/bacterias.csv")
```

```python

viruses = {
    bacteria: metadata[metadata["Isolation Host (beware inconsistent and nonsense values)"].str.contains(bacteria)]["Accession"].unique()
    for bacteria in bacterias["bacteria"]
}
```

```python
selected_viruses = {
    bacteria: random.choices(viruses[bacteria], k=N_PHAGES)
    for bacteria in viruses
}
```

```python
selected_viruses_df = pd.DataFrame(selected_viruses)
selected_viruses_df.to_csv("../config/selected_virus.csv", index=False)
```

```python
selected_viruses_list = []
for values in selected_viruses.values():
    selected_viruses_list.extend(values)
df = pd.DataFrame(dict(viruses=np.array(selected_viruses_list)))
df.to_csv("../config/viruses.list", index=False)
```
